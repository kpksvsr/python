```python
from scipy import stats
def chi_sq_test_cat(df1,col1,col2,lvl_of_sig=0.05):
    ct1 = pd.crosstab(df1[col1], df1[col2],margins=False)
    p_value = stats.chi2_contingency(ct1.values)[1]
    print('P-Value is:{}'.format(p_value))
    if(p_value < lvl_of_sig):
        print('The attributes are dependent')
    else:
        print('The attributes are independent')
```

```python
def merge_levels(df1,colname):
    first_run=1
    for v1 in pd.unique(df1[colname]):
        #print('current category is:{}'.format(v1))
        if(len(df1[df1[colname] == v1]) <= 500):
            if(first_run == 1):
                #print('first run')
                first_run=0
                #df1[colname] = df1[colname].apply(str)
                #print('add others category')
                df1[colname].cat.add_categories('others',inplace=True)
                #print('added category successfully')
            #print('set new category to existing')
            df1.loc[df1[colname] == v1,colname] = 'others'
            #print('done setting new category')
    df1[colname].cat.remove_unused_categories(inplace=True)
```

```python
def merge_into_existing(df1,colname,src,tgt):
    df1.loc[df1[colname] == src,colname] = tgt
    df1[colname].cat.remove_unused_categories(inplace=True)
```

```python
def total_acc(cnf_dframe):
    diag = 0;
    total_sum = 0;
    for i in range(len(cnf_dframe)):
        for j in range(len(cnf_dframe.columns)):
            total_sum += cnf_dframe.iloc[i,j]
            if(i == j):
                diag += cnf_dframe.iloc[i,j]
    return (diag/total_sum)*100
                
```
